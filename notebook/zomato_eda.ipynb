```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pltpip 
import seaborn as sns
```

```python
df = pd.read_csv("../data/zomato-data-.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
df.isnull().sum()
```

```python
df["rate"] = df["rate"].astype(str)
df["rate"] = df["rate"].str.replace("/5", "", regex=False)
df["rate"] = pd.to_numeric(df["rate"], errors="coerce")
```

```python
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].astype(str)
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].str.replace(",", "")
df["approx_cost(for two people)"] = pd.to_numeric(df["approx_cost(for two people)"], errors="coerce")
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["rate"])
```

```python
df["rate"].mean()
df["approx_cost(for two people)"].mean()
```

```python
df["listed_in(type)"].value_counts()
```

```python
df.groupby("online_order")["rate"].mean()
```

```python
df.groupby("book_table")["rate"].mean()
```

```python
df[["votes","rate"]].corr()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["rate"], bins=10, kde=True)
plt.title("Rating Distribution")
plt.show()
```
